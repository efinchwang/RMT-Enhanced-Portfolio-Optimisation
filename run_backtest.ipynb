```python
from strategy import StaticPCAStrategy
from backtester import Backtester
```

```python
tickers = ['XLE', 'XLF', 'XLK', 'XLI', 'XLV', 'XLP', 'XLY', 'XLB', 'XLU']
start_date = "2000-01-01"
end_date = "2025-09-01"
lookback_period = 63
```

```python
strategy = StaticPCAStrategy(
    ticker_list = tickers,
    start = start_date,
    end = end_date,
    rolling_window = lookback_period
)

data = strategy.import_data()

backtester = Backtester(strategy = strategy, data = data)
```

```python
backtester.run()
```

```python
performance = backtester.calculate_performance_metrics()

print("--- Performance Metrics ---")
print("Annualised Return:", performance["Annualised Return"])
print("Sharpe Ratio:", performance["Sharpe Ratio"])
print("Maximum Drawdown:", performance["Maximum Drawdown"])
```
